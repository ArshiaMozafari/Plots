```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def make_plot(
    x=None, y=None, title=None, xlabel=None, ylabel=None, 
    func=None, xlim: tuple | None = None, ylim: tuple | None = None
):
    plt.figure(figsize=(7, 5))

    if x is not None and y is not None:
        plt.scatter(x, y, color='red', label="Data points")

    if xlim is not None:
        xs = np.linspace(xlim[0], xlim[1], 500)
        plt.xlim(xlim)
    elif x is not None:
        xs = np.linspace(min(x), max(x), 500)
    else:
        xs = np.linspace(-10, 10, 500)

    if ylim is not None:
        plt.ylim(ylim)

    if func is not None:
        try:
            ys = func(xs)
            plt.plot(xs, ys, color="blue", label="Line/Curve")
        except Exception as e:
            raise ValueError(f"Could not evaluate function on xs: {e}")

    if title: plt.title(title)
    if xlabel: plt.xlabel(xlabel)
    if ylabel: plt.ylabel(ylabel)

    plt.minorticks_on()
    plt.grid(True, which='major', linestyle='-', linewidth=0.9)
    plt.grid(True, which='minor', linestyle='--', linewidth=0.5, alpha=0.7)

    handles, labels = plt.gca().get_legend_handles_labels()
    if handles:
        plt.legend()

    plt.show()
```

```python
#example
make_plot( np.array([1,2,3,4]), np.array([1, 5, 10, 15]),  title = 'My plot', xlabel = 'input', ylabel= 'output', func = lambda x:  x**2, xlim = (0,10), ylim = (0, 20))
```
